```python
import requests as req
import os   #for 
import json #json 
from bs4 import BeautifulSoup as bs
from pprint import pprint
#  User-Agent
'''
:
https://pypi.org/project/fake-useragent/
'''

from fake_useragent import UserAgent 
ua = UserAgent()

#  
folderPath = 'line_stickers'
if not os.path.exists(folderPath):
    os.makedirs(folderPath)

    
#
listLineStickers = []

# 
my_headers = {
    'user-agent':ua.random
}

#  LINE 
url = 'https://store.line.me/stickershop/product/25592993/zh-Hant'

#  GET 
res = req.get(url,headers = my_headers)
soup = bs(res.text,'lxml')

'''
:
1. soup.select()list BeautifulSoup ResultSet
2. soup.select_one()BeautifulSoup Result
'''
#  li  (list )
li_elements = soup.select('ul.mdCMN09Ul.FnStickerList > li.mdCMN09Li.FnStickerPreviewItem.static-sticker ')#ctrl+f:

print(li_elements)##dict

#  li  data-preview 
for li in li_elements:
    #  data-preview ()
    strjson = li['data-preview']

    #()
    obj = json.loads(strjson) # li.get("data-preview")
    
    #  list 
    listLineStickers.append(obj)


pprint(listLineStickers)


# 
'''
:
$ curl "https://stickershop.line-scdn.net/stickershop/v1/sticker/541202232/iPhone/sticker@2x.png?v=4" -o ./test.png
'''
for obj in listLineStickers:
    os.system(f"curl {obj['staticUrl']} -o {folderPath}/{obj['id']}.png")
```
